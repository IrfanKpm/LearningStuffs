```python
import seaborn as sns
```

```python
sns.get_dataset_names()  # dataset ava
```

```python
tips = sns.load_dataset("tips")
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset("iris")
```

```python
tips
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips)
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day")
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size")
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size",palette="YlGnBu")
```

```python
sns.histplot(tips['tip'])
```

```python
sns.histplot(tips['tip'],kde=True)
```

```python
# you can specify the number of bins to control the granularity of your visualization
sns.histplot(tips['tip'],kde=True,bins=30)
```

```python
sns.barplot(x="sex",y="tip",data=tips,hue='sex',palette="YlGnBu")
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',palette="YlGnBu")
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex',palette="YlGnBu",dodge=True) # saperate sex 
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='kde',fill=True,cmap="YlGnBu")  # cmap :: colormap
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='hex',cmap="YlGnBu")
```
